```python
# PART 2 Data Cleaning
# 1 Handling Missing Values

# Missing values were check on Part 1 Point 2 

# For numerical columns with missing values, replace them with the column mean
# For products
products['weight'] = products ['weight'].fillna(products['weight'].mean())
# For stores
stores['size_sqft'] = stores ['size_sqft'].fillna(stores['size_sqft'].mean()) 
```

```python
# For categorical columns with missing values, replace them with the most frequent value
# For customers
customers['email'] = customers ['email'].fillna(customers['email'].mode()[0])
customers['phone'] = customers ['phone'].fillna(customers['phone'].mode()[0])
customers['zip_code'] = customers ['zip_code'].fillna(customers['zip_code'].mode()[0])
# For sales
sales['store_id'] = sales ['store_id'].fillna(sales['store_id'].mode()[0])
sales['payment_method'] = sales ['payment_method'].fillna(sales['payment_method'].mode()[0])
# For inventory
inventory['reorder_level'] = inventory ['reorder_level'].fillna(inventory['reorder_level'].mode()[0])
```

```python
# For date columns with missing values, use forward fill or backward fill as appropriate
# For customers
customers['registration_date'] = pd.to_datetime(customers['registration_date'])
customers['registration_date'] = customers['registration_date'].fillna(method='ffill')
# For stores
stores['opening_date'] = pd.to_datetime(stores['opening_date'])
stores['opening_date'] = stores['opening_date'].fillna(method='ffill')
# For sales
sales['date'] = pd.to_datetime(sales['date'])
sales['date'] = sales['date'].fillna(method='ffill') 
# For inventory
inventory['last_restock_date'] = pd.to_datetime(inventory['last_restock_date'])
inventory['last_restock_date'] = inventory['last_restock_date'].fillna(method='ffill')
```

```python
# 2 Removing Duplicates
# Customers
customers.duplicated(subset=['customer_id' , 'first_name' , 'last_name']).sum()
```

```python
dupes = customers[customers.duplicated(subset=['customer_id', 'first_name', 'last_name'], keep=False)]
dupes.sort_values(by=['customer_id'])
```

```python
#To identify duplicates, I focused on key columns that should be unique, such as customer_id and product_id. 
#These identifiers are meant to represent individual entities in the dataset, so any duplicates in them could indicate data entry errors 
#or inconsistencies. On the other hand, I intentionally ignored columns like state or city, since those values can be repeated 
#across many records without affecting data integrity. My goal was to isolate and handle only the duplicates that could impact the uniqueness
#and reliability of the core data.
```
